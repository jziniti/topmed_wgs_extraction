```python
import os
import pandas as pd

WORKDIR = '/proj/regeps/regep00/studies/TopMed/data/dna/whole_genome/TopMed/data/freezes/freeze.10.cdnm'
SK38MC_MANIFEST_PATH = 'code/COPD_lost_samples/SK-38MC_NS Controls_NWD.xlsx'
UNASSIGNED_PATH = 'code/COPD_lost_samples/output/remain_unassigned.txt'
os.chdir(WORKDIR)
```

```python
sk38 = pd.read_excel(SK38MC_MANIFEST_PATH)
sk38
```

```python
unassigned = pd.read_csv(UNASSIGNED_PATH, names=('NWDID',))
unassigned
```

```python
unassigned_from_sk38 = unassigned.merge(sk38, how='left', left_on='NWDID', right_on='Alias.1')
unassigned_from_sk38
```

```python
unassigned_after_sk38 = unassigned_from_sk38[unassigned_from_sk38['Alias.1'].isnull()]
unassigned_after_sk38.to_csv('code/COPD_lost_samples/output/unassigned_after_sk38.txt', index=False, columns=("NWDID",))
unassigned_after_sk38
```
