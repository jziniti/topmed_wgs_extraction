```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_sexcheck = pd.read_csv("/proj/regeps/regep00/studies/TopMed/data/dna/whole_genome/TopMed/data/freezes/freeze.10.cdnm/tmp/GECOPD_annotated_plink_merged.sexcheck", delim_whitespace=True)
gz = pd.read_csv("/udd/remhc/Work/Wgs/TOPMed_metrics_2021-05-20-111439.tab.gz", compression='gzip', delim_whitespace=True)

```

```python
df_sexcheck.columns = ['FID', "SAMPLE_ID", "PEDSEX", "SNPSEX", 'STATUS', 'F']
df_sexcheck
```

```python
gz
```

```python
data = pd.merge(df_sexcheck, gz, on= 'SAMPLE_ID')
data.dropna
data
```

```python
plt.scatter(data['F'], data['SNPSEX'],  alpha=0.3)
```

```python
len(data.query("SNPSEX == 0")['SNPSEX'])
```

```python
plt.scatter(data['BAM_DATE'], data['SNPSEX'],  alpha=0.3)
plt.rcParams["figure.figsize"] = (30,5)
plt.xticks(rotation = 90) 
plt.show()
```

```python
snp_zero = data.query('SNPSEX == 0')
sns.displot(snp_zero['BAM_DATE'], kde=True)
plt.title('Samples distribution by days')
plt.xticks(rotation = 90) 
plt.show()
```

```python
snp_notzero = data.query('SNPSEX != 0')
sns.displot(snp_zero['BAM_DATE'], kde=True)
plt.title('Samples distribution by days')
plt.xticks(rotation = 90) 
plt.show()
```
