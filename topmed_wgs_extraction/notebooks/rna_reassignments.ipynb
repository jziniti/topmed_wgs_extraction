```python
import pandas as pd
import os

WORKDIR = '/proj/regeps/regep00/studies/TopMed/data/dna/whole_genome/TopMed/data/freezes/freeze.10.cdnm'
os.chdir(WORKDIR)
```

```python
MASTERFILE = '/proj/edith/regeps/regep00/studies/COPDGene/data/rnaseq/BLOOD/archivalroot/Masterfile/master.file.freeze4.txt'
masterfile = pd.read_csv(MASTERFILE, sep='\t')
masterfile
```

```python
reassignments = masterfile[masterfile['COPDgeneID'] != masterfile['actual_id']]
reassignments = reassignments[reassignments['final.analyzed'] == 1]
reassignments = reassignments[['sample_id', 'COPDgeneID', 'actual_id']]
reassignments 
```

```python
ALIASES = '/proj/regeps/regep00/studies/COPDGene/metadata/topmed_aliases.txt'
aliases = pd.read_csv(ALIASES)
aliases = aliases[aliases['ALIASTYPE'] == 'Pheno Id']
aliases = aliases.drop_duplicates()
aliases
```

```python
reassignments = reassignments.merge(aliases, left_on='COPDgeneID', right_on='ALIASID', how='left')
reassignments = reassignments.merge(aliases, left_on='actual_id', right_on='ALIASID', how='left')
reassignments = reassignments[['sample_id', 'COPDgeneID', 'actual_id', 'S_SUBJECTID_x', 'S_SUBJECTID_y']]
reassignments = reassignments.rename(columns={'S_SUBJECTID_x': 'reassign_from', 'S_SUBJECTID_y': 'reassign_to'})
reassignments.drop_duplicates(inplace=True)
reassignments.to_csv(f'{WORKDIR}/freeze10qc_rna_reassignments.csv', index=False)
reassignments
```
