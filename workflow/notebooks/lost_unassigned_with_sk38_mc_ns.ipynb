```python
import os
import pandas as pd

WORKDIR = '/proj/regeps/regep00/studies/TopMed/data/dna/whole_genome/TopMed/data/freezes/freeze.10.cdnm'
SK38MC_MANIFEST_PATH = 'code/COPD_lost_samples/SK-38MC_NS Controls_NWD.xlsx'
UNASSIGNED_PATH = 'code/COPD_lost_samples/output/remain_unassigned.txt'
FULL_UPDATED_012419 = 'code/COPD_lost_samples/metadata/COPD WGS Subject List_FULL_Updated 012419.xlsx'
MISSING_170726 = 'code/COPD_lost_samples/metadata/TopMed_Sample_Status_170726-missing (1).csv'
os.chdir(WORKDIR)
```

```python
sk38 = pd.read_excel(SK38MC_MANIFEST_PATH)
sk38
```

```python
unassigned = pd.read_csv(UNASSIGNED_PATH, names=('NWDID',))
unassigned
```

```python
unassigned_from_sk38 = unassigned.merge(sk38, how='left', left_on='NWDID', right_on='Alias.1')
assigned_to_study = unassigned_from_sk38[unassigned_from_sk38['Alias.1'].notnull()]
assigned_to_study.to_csv('code/COPD_lost_samples/output/assigned_from_sk38.txt', index=False, columns=('NWDID',))
assigned_to_study
```

```python
unassigned_after_sk38 = unassigned_from_sk38[unassigned_from_sk38['Alias.1'].isnull()]
unassigned_after_sk38.to_csv('code/COPD_lost_samples/output/unassigned_after_sk38.txt', index=False, columns=("NWDID",))
unassigned_after_sk38
```

```python
full_updated = pd.read_excel(FULL_UPDATED_012419)
full_updated
```

```python
assigned_by_full = unassigned_after_sk38.merge(full_updated, left_on='NWDID', right_on='NWD ID')
assigned_by_full
```

```python
missing = pd.read_csv(MISSING_170726)
assigned_from_missing = unassigned_after_sk38.merge(missing, left_on='NWDID', right_on='NWD_ID')
assigned_from_missing
```
