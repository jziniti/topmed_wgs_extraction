```python
f8_annotated['flags'].fillna('', inplace=True)
cross_subject_matches = f8_annotated[f8_annotated['flags'].str.contains('cross_subject_match')]
cross_subject_matches
```

```python
# cross_subject_matches.to_csv('tmp/freeze.8_cross_subject_matches.csv', index=False)
```

```python
## Integrate concordance data for these samples ...
c_data = pd.read_csv('/proj/regeps/regep00/studies/TopMed/data/dna/whole_genome/TopMed/data/freezes/freeze.10.cdnm/multiomics/GECOPD/c_data.build_sample_groups')
c_data = c_data[(c_data['expected'] == 0) & (c_data['observed'] == 1)]
c_data
```

```python
bad_matches_with_c_data = cross_subject_matches.merge(c_data, left_on='SampleLabel', right_on='ID1')
bad_matches_with_c_data = bad_matches_with_c_data[~bad_matches_with_c_data['ID2'].str.startswith('S-')]
bad_matches_with_c_data = bad_matches_with_c_data[~bad_matches_with_c_data['ID2'].str.startswith('TOE')]
bad_matches_with_c_data.drop(columns=['S_SUBJECTID_participant', 'c_group_participant'], inplace=True)
bad_matches_with_c_data.to_csv('tmp/freeze.8_cross_subject_matches.csv', index=False, columns=['SampleAlias', 'ExpectedGender', 'ObservedGender', 'group_id', 'N', 'ALIASID', 'ID2', 'Concord', 'flags'])
bad_matches_with_c_data[['SampleAlias', 'ExpectedGender', 'ObservedGender', 'group_id', 'N', 'ALIASID', 'ID2', 'Concord', 'flags']].sort_values(['N', 'group_id'])
```
