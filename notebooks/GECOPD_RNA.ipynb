```python
batch_files = pd.read_csv('conf/all_rna_batches.txt', names=['bed'])
batch_files['fam'] = batch_files['bed'].str.replace('.bed', '.fam')
batch_files
```

```python
all_rna_samples = []
for fd in batch_files['fam'].values:
    fam = pd.read_csv(fd, names=('FID', 'IID', 'DID', 'MID', 'SEX', 'AFF'), delim_whitespace=True)
    fam = fam[fam['IID'].str.startswith('S-')]
    all_rna_samples.append(fam)
all_rna_samples = pd.concat(all_rna_samples)
all_rna_samples
```

```python
annotated = all_rna_samples.merge(masterfile, left_on='IID', right_on='BAM_file', how='left')
annotated[annotated['sample_id'].isna()]
```
