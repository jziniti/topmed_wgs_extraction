```python
import pandas as pd
import os

os.chdir('/proj/regeps/regep00/studies/TopMed/data/dna/whole_genome/TopMed/data/freezes/freeze.10.cdnm')
```

```python
ROSETTA_PATH = '/proj/regeps/regep00/studies/COPDGene/analyses/rejpz/Rosetta/Rosetta_stone_annotated.txt'

rosetta = pd.read_csv(ROSETTA_PATH, delimiter='\t')
rosetta
```

```python
### These are the subjects that Carla flagged as plate103
flagged = rosetta[rosetta['comments_dcc'].str.contains(' 103')]
flagged
```

```python
flagged_s = flagged[['Subjectid_Channing', 'S_Sampleid', 'Topmed_ID', 'Collection', 'sid_Channing', 'sid_dcc', 'sid_match', 'Channing_action', 'comments_dcc', 'issue_flag', 'Qc_Status']]
flagged_s
```

```python
manifest = pd.read_csv('multiomics/GECOPD/ANNOTATED_MANIFEST.csv')
manifest
```

```python
### Filter the manifest by the subjects that are on Carla's list
filtered_manifest = manifest[manifest['S_SUBJECTID'].isin(flagged_s['Subjectid_Channing'])]
filtered_manifest
```

```python
merged_dataset = flagged_s.merge(filtered_manifest, left_on='Subjectid_Channing', right_on='S_SUBJECTID', how='left')
merged_dataset = merged_dataset.drop(columns=['is_reference', 'ExpectedGender', 'ObservedGender', 'keep_drop', 'SampleLabel', 'SampleTypeCode'])
merged_dataset
```

```python
### Read in the concordance information between freeze.10 and the unified dataset
concordance = pd.read_csv('tmp/GECOPD_reference_concordance.con', delimiter='\t')
concordance
```

```python
### Merge in the concordance information
big_kahuna = merged_dataset.merge(concordance, how='left', left_on='SampleAlias', right_on='ID1')
big_kahuna = big_kahuna.drop(columns=['FID2', 'FID1', 'N_IBS0', 'N_IBS1', 'N_IBS2', 'HomConc', 'HetConc',])
big_kahuna
```

```python
big_kahuna.to_csv('tmp/plate103_summary_20210928.csv')
```
