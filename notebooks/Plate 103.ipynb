```python
plate_7962 = pd.read_csv('../../plate103/PL-01007962_v2.csv', usecols=['SUBJECTID', 'ALIASES'])
plate_7962
```

```python
plate_7365 = pd.read_csv('../../plate103/PL-01007365_v2.csv', usecols=['SUBJECTID', 'ALIASES'])
plate_7365
```

```python
plate_7396 = pd.read_csv('../../plate103/PL-01007396_v2.csv', usecols=['SUBJECTID', 'ALIASES'])
plate_7396
```

```python
all_plates = pd.concat([plate_7365, plate_7396, plate_7962]).drop_duplicates()
all_plates
```

```python
grouped = manifest[['S_SUBJECTID', 'status']].groupby('S_SUBJECTID')['status'].transform(lambda x: ';'.join(list(set(x))))
grouped
```

```python
multiomics_results = manifest[['S_SUBJECTID', 'SampleAlias', 'flags']]
# multiomics_results['ALIASES'] = multiomics_results.groupby('S_SUBJECTID')['SampleAlias'].transform(lambda x: '|'.join(list(set(x))))
multiomics_results
```

```python
merged = all_plates.merge(multiomics_results, left_on='SUBJECTID', right_on='S_SUBJECTID', how='left')
merged = merged.drop(columns=['S_SUBJECTID'])
merged
```

```python
subjects = merged['SUBJECTID'].unique()
for subject in subjects:
    all_rows = merged[merged['SUBJECTID'] == subject]
    # print(all_rows['flags'])
    for row in all_rows.iterrows():
        print(f'{subject}, {row[1]["ALIASES"]}, {row[1]["flags"]}')
    print('OK')
    
```
