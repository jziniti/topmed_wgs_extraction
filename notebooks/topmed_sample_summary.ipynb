```python
import pandas as pd
import os
```

```python
WORKDIR = '/proj/regeps/regep00/studies/TopMed/data/dna/whole_genome/TopMed/data/freezes/freeze.10.cdnm'
IMISS_FILE = 'tmp/GECOPD_annotated_plink_merged.imiss'
SEXCHECK_FILE = 'tmp/GECOPD_annotated_plink_merged.sexcheck'
MULTIOMICS_FILE = 'multiomics/GECOPD/ANNOTATED_MANIFEST.csv'
HET_FILE = 'tmp/GECOPD_annotated_plink_merged.het'
C_DATA_FILE = 'multiomics/GECOPD/c_data.build_sample_groups'
```

```python
os.chdir(WORKDIR)
```

```python
het = pd.read_csv(HET_FILE, delim_whitespace=True, usecols=['IID', 'F'])
het
```

```python
imiss = pd.read_csv(IMISS_FILE, delim_whitespace=True, usecols=['IID', 'F_MISS'])
imiss
```

```python
sexcheck = pd.read_csv(SEXCHECK_FILE, delim_whitespace=True, usecols=['IID', 'SNPSEX', 'F'])
sexcheck = sexcheck.rename(columns={'F': 'XHomFStat'})
sexcheck
```

```python
multiomics = pd.read_csv(MULTIOMICS_FILE)
multiomics = multiomics.merge(het, left_on='SampleAlias', right_on='IID', how='left')
multiomics = multiomics.merge(imiss, left_on='SampleAlias', right_on='IID', how='left')
multiomics = multiomics.merge(sexcheck, left_on='SampleAlias', right_on='IID', how='left')
multiomics
```

```python
wgs_samples = multiomics[multiomics['dataset_id'] == 0]
wgs_samples
```

```python
wgs_samples[(wgs_samples['F'] > 0.2) | (wgs_samples['F'] < -0.2)]
```

```python
wgs_samples[(wgs_samples['SNPSEX'] == 0)]
```

```python
wgs_samples[(wgs_samples['ExpectedGender'] == 1) & (wgs_samples['XHomFStat'] < 0.8)]
```

```python
wgs_samples[(wgs_samples['ExpectedGender'] == 2) & (wgs_samples['XHomFStat'] > 0.2)]
```

```python
wgs_samples[(wgs_samples['F_MISS'] > 0.1)]
```

```python
wgs_samples['fail_heterozygosity'] = (wgs_samples['F'] > 0.2) | (wgs_samples['F'] < -0.2)
wgs_samples['fail_sexcheck'] = wgs_samples['SNPSEX'] == 0
wgs_samples['fail_missing'] = wgs_samples['F_MISS'] > 0.1
wgs_samples
```

```python
flagged = wgs_samples[(wgs_samples['fail_sexcheck'] == True) | (wgs_samples['fail_heterozygosity'] == True) | (wgs_samples['fail_missing'] == True)]
flagged[['SampleAlias', 'fail_heterozygosity', 'F', 'fail_sexcheck', 'XHomFStat', 'fail_missing', 'F_MISS', 'flags', 'action']]
flagged.to_csv('multiomics/GECOPD/flagged_samples.csv')
```

```python
p0 = pd.read_csv(PLATE_MANIFESTS[0])
p1 = pd.read_csv(PLATE_MANIFESTS[1])
p2 = pd.read_csv(PLATE_MANIFESTS[2])
plate_manifests = pd.concat([p0, p1, p2])
plate_manifests
```
