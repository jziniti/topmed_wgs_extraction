### Set up the database connection

```python
import pandas as pd

from sqlalchemy import create_engine, MetaData, Table
from sqlalchemy.orm import sessionmaker

sapphire8_md = MetaData(schema='SAPPHIRE8')
engine = create_engine('oracle://sapphire_r:sapphire_r0@labvprd')
Session = sessionmaker(bind=engine)
session = Session()

SDIAlias = Table('SDIALIAS', sapphire8_md, autoload=True, autoload_with=engine)
print(SDIAlias.columns)
```

### NWDIDs

```python
nwdids = pd.read_sql(session.query(SDIAlias).where(SDIAlias.c.aliasid.like('NWD%')).statement, engine)
nwdids
```

### TOEIDs

```python
toeids = pd.read_sql(session.query(SDIAlias).where(SDIAlias.c.aliasid.like('TOE%')).statement, engine)
toeids
```

### TORIDs

```python
torids = pd.read_sql(session.query(SDIAlias).where(SDIAlias.c.aliasid.like('TOR%')).statement, engine)
torids
```

## Now: All TOPMed samples (NWD, TOE, TOR)

```python
topmed_samples = pd.concat([nwdids, toeids, torids]).reset_index()
topmed_samples['sample_type'] = topmed_samples['aliasid'].str[:3]
topmed_samples
```
